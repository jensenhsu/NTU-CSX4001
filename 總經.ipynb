# 讀入資料

```python
import pandas as pd
job_info = pd.read_csv("jobseeking.csv")
job_info_0 = job_info.drop("Unnamed: 2", axis = 1)
job_info_0.head()
```

# 繪圖

```python
#以全體的資料看

import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"] = ["simhei"] 


plt.plot(job_info_0["duration"])
plt.title("週期趨勢圖" , color = "r" , size = 20)
plt.ylabel("失業週數" , color = "r" , size = 20)
plt.show()
```

```python
#只看一半

plt.rcParams["font.sans-serif"] = ["simhei"] 

half_duration_data = job_info_0["duration"][1: 167]
plt.plot(half_duration_data)
plt.title("週期趨勢圖" , color = "r" , size = 20)
plt.ylabel("失業週數" , color = "r" , size = 20)
plt.show()
```

# 失業的週期：
根據摩擦性失業，摩擦性失業即是勞工在尋找工作期間所造成的失業，而隨著失業的週數越拉越長，政府為了降低摩擦性失業來減少自然失業率，就可能會想辦法製造工作機會，使失業的人口能夠被吸收，失業週數就會下降

# 過去三年的平均

```python
past_three_year = job_info_0.iloc[288 : 321]
past_three_year.head()
```

```python
past_three_year_duration = past_three_year["duration"]
past_three_year_mean = past_three_year_duration.mean()
past_three_year_mean
```

```python
print("過去三年的平均失業週數：{: .2f}".format(past_three_year_mean))
```

## 若失業30週
由以上的計算可以知道過去三年的平均失業週數大約為26週，若畢業後30周都還沒有找到工作，以過去三年的平均水準而言，超過了4週，所以父母應該擔心，因為此人表現得比平均還要差
