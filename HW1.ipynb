```python
import pandas as pd
import matplotlib.pyplot as plt


file = pd.read_csv("gapminder.csv")

file.head()
```

```python
plt.hist(file["gdpPercap"],bins = 100)
plt.show
```

```python
import seaborn as sns 
sns.lmplot("gdpPercap", "lifeExp", data = file, hue = "continent",fit_reg=False)
plt.show()
```

```python
taiwan = file[file["country"] == "Taiwan"]
taiwan_year = taiwan["year"]
taiwan_gdpPercap = taiwan["gdpPercap"]
plt.plot(taiwan_year , taiwan_gdpPercap)
plt.xlabel("year")
plt.ylabel("gdpPercap")
plt.show
```

```python
taiwan_lifeExp = taiwan["lifeExp"]
plt.plot(taiwan_year, taiwan_lifeExp)

plt.xlabel("year")
plt.ylabel("lifeyear")
plt.show
```
