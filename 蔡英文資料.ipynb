## 1.建計算推文數的函數

```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
from pyquery import PyQuery as pq
from time import sleep
```

```python
April_Data_0 = pd.read_csv("April_Gossiping_Data.csv")
April_Data_0.head()
April_Data = April_Data_0.rename(columns = {"Push_Count":"Comment_Count"})

April_Data.head()
```

```python
def get_Push_Boo_Count(URL):
    '''單篇文章的推文數'''
    
    res = requests.get(url = 'https://www.ptt.cc' + URL, cookies = {'over18':'1'})
    soup = BeautifulSoup(res.text, 'html.parser')

        
    # 總推文數計算
    com = soup.find_all('div', 'push')
    count_1 = 0

    for elem in com:
        if elem.find('span', 'hl push-tag'):
            count_1 += 1
    
    num_1 = [count_1]
    
    
    
    #總噓文數計算
    com = soup.find_all('div', 'push')
    types = [ ]
    count_2 = 0

    for elem in com:
        if elem.find('span', 'f1 hl push-tag'):
            types.append(elem.find("span" , "f1 hl push-tag").text.strip())
            
    for i in types:
        if i == "噓":
            count_2 +=1
    
    num_2 = [count_2]
    
    info = pd.DataFrame({"Push_Count":num_1 , "Boo_Count":num_2})
    
    
    
    return info
```

## 2.篩選內文中有蔡英文的文章

```python
#篩選有蔡英文的文章

bool = April_Data["Content"].str.contains('蔡英文')
Tsai_data = April_Data[bool]
Tsai_data.head()
```

## 3.計算有提到蔡英文的文章推噓文數量

```python
link_ser = Tsai_data["Link"]
len(link_ser)
```

```python
#前1000篇

count_push_boo = pd.DataFrame(columns=['Push_Count','Boo_Count'])
count = 0

if __name__ == '__main__':
    for link in link_ser[0:1000]:
        df = get_Push_Boo_Count(link)
        count_push_boo = pd.concat([count_push_boo, df], ignore_index=True, sort=False)
        count += 1
        
        if count % 100 == 0:
            sleep(5)
            
            
count_push_boo.head()
```

```python
count_push_boo.to_csv('count_push_boo_0.csv', columns=['Push_Count', 'Boo_Count'])
```

```python
#1000到2000篇

count_push_boo_1 = pd.DataFrame(columns=['Push_Count','Boo_Count'])
count = 0

if __name__ == '__main__':
    for link in link_ser[1000:2000]:
        df = get_Push_Boo_Count(link)
        count_push_boo_1 = pd.concat([count_push_boo_1, df], ignore_index=True, sort=False)
        count += 1
        
        if count % 100 == 0:
            sleep(5)
            
            
count_push_boo_1.head()
```

```python
count_push_boo_1.to_csv('count_push_boo_1.csv', columns=['Push_Count', 'Boo_Count'])
```

```python
#2000到3700篇

count_push_boo_2 = pd.DataFrame(columns=['Push_Count','Boo_Count'])
count = 0

if __name__ == '__main__':
    for link in link_ser[2000:3700]:
        df = get_Push_Boo_Count(link)
        count_push_boo_2 = pd.concat([count_push_boo_2, df], ignore_index=True, sort=False)
        count += 1
        
        if count % 100 == 0:
            sleep(5)
            
            
count_push_boo_2.head()
```

```python
count_push_boo_0 = pd.read_csv("count_push_boo_0.csv")
count_push_boo_1 = pd.read_csv("count_push_boo_1.csv")
```

```python
tsai_push_boo_0 = pd.concat([count_push_boo_0 , count_push_boo_1 , count_push_boo_2] , axis = 0 , ignore_index = True , sort = False)
tsai_push_boo = tsai_push_boo_0.drop(columns = ["Unnamed: 0"])
tsai_push_boo.head()
```

## 4.合併兩組DataFrame

```python
tsai_info = pd.concat([Tsai_data , tsai_push_boo] , axis = 1 , ignore_index = True)
tsai_info_remove = tsai_info.drop(columns = [0 , 1 , 2])

Tsai_April_info = tsai_info_remove.rename(columns = {3:"Date" , 4:"Push" , 5:"Title",6:"Author",7:"Link",8:"Comment_Count",9:"Content",10:"Push_Count",11:"Boo_Count"})

Tsai_April_info
```
